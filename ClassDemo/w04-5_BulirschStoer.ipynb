### Bulirsch-Stoer 

A good choice for smooth, non-singular, non-stiff systems of ODE's

The midpoint method (aka second-order Runge-Kutta) used a forward Euler step of $h/2$ to the midpoint of an interval of $h$, used the result to compute the slope at that point, and then used that slope to advance over the whole interval
\begin{align}
k_1 &= h f(x_n, y_n) \\
k_2 &= h f(x_n+\frac{1}{2} h, y_n + \frac{1}{2}k_1) \\
y_{n+1} &= y_n + k_2
\end{align}

The *modified midpoint method* advances the solution over a timestep $H$ by dividing that interval
into $n$ sub-intervals with timestep $h = H/n$. The first two steps are the midpoint method, followed by
the three-point recursion
\begin{equation*}
\begin{split}
z_0 & = y(x) \\
z_1 & = z_0 + h f(x, z_0) \\
z_{m+1} &= z_{m-1} + 2 h f(x+mh, z_m), \quad m=1,2,\dots,n-1 \\
y(x+H) \sim y_n &= \frac{1}{2}\left[z_n + z_{n-1} + h f(x+H, z_n)\right]
\end{split}
\end{equation*}

A simple implementation in our by now usual style is:

```python
def modMid(t, y, H, f, *P):
    k = 4
    
    h = H / k
    z = np.zeros((k+1,len(y)))
    z[0,:] = y
    z[1,:] = z[0,:] + h*f(t, z[0,:], *P)
    for i in range(2,k+1):
        z[i,:] = z[i-2,:] + 2*h*f(t+i*h,z[i-1,:], *P)
    return 0.5*(z[k,:]+z[k-1,:]+h*f(t+H,z[k,:], *P))
```

For comparison, we'll return to our old standby, RK4

```python
def RK4(t, y, h, RHS, *P):
    """
    Implements a single step of a fourth-order, explicit Runge-Kutta scheme
    """
    thalf = t + 0.5*h
    k1 = h * RHS(t, y, *P)
    k2 = h * RHS(thalf, y + 0.5*k1, *P)
    k3 = h * RHS(thalf, y + 0.5*k2, *P)
    k4 = h * RHS(t + h, y + k3, *P)
    return y + (k1 + 2*k2 + 2*k3 + k4)/6
```

This is the simple driver from before

```python
def odeSolve(RHS, interval, y0, h, method, *P):
    """
    ODE driver with constant step-size, allowing systems of ODE's
    """
    
    t0, tmax = interval
    
    # make array of times and find length of array
    t = np.arange(t0,tmax+h,h)
    ntimes,  = t.shape
    
    # find out if we are solving a scalar ODE or a system of ODEs, and allocate space accordingly
    
    if type(y0) in [int, float]:  # check if primitive type -- means only one eqn
        neqn = 1
        y = np.zeros( ntimes )
    else:                         # otherwise assume a numpy array -- a system of more than one eqn
        neqn, = y0.shape
        y = np.zeros( (ntimes, neqn) )

    # set first element of solution to initial conditions (possibly a vector) 
    y[0] = y0

    # time marches on...
    for i in range(0,ntimes-1):
        y[i+1] = method(t[i], y[i], h, RHS, *P)

    return t,y
```

As an example, we'll use the Lotke-Volterra system (aka preditor-prey equations) from the first set of examples. Here is the RHS function we used:

```python
def lvRHS(t, y, *P):
    # Lotke-Volterra system RHS
    
    # unpack the parameters from the array P
    alpha, beta, gamma, delta = P

    # make temporary variables with rabbit and fox populations
    R = y[0]
    F = y[1]
    
    # LV system
    dRdt = alpha * R     -  beta * R * F
    dFdt = delta * R * F - gamma *     F
    
    # return an array of derivatives with same order as input vector
    return np.array([ dRdt, dFdt ])
```

First start with RK4 and a very small timestep as a "gold standard" (not that RK4 is "better" in any way, just that we have it, know it works, and with a sufficiently small stepsize is likely to be quite accurate if inefficient).

```python
alpha = 1/2
beta = 1/80
gamma = 1/4
delta = 1/200

h = 0.02
t0 = 0.0
y0 = np.array([ 30, 10 ])
tmax = 50

# call the solver
trk4, yrk4 = odeSolve(lvRHS, (t0, tmax), y0, h, RK4, alpha, beta, gamma, delta)

fig,ax = plt.subplots(1,2)
ax[0].plot(trk4,yrk4[:,0],'b', lw=0.5, label='prey')
ax[0].plot(trk4,yrk4[:,1],'r', lw=0.5, label='preditor')
ax[0].set_xlabel('time')
ax[0].set_ylabel('population')
ax[0].legend()

ax[1].plot(yrk4[:,0],yrk4[:,1],'g', lw=0.5)
ax[1].set_xlabel('prey')
ax[1].set_ylabel('preditor')

plt.tight_layout()
plt.show()
```

As you can see from the solution, the modified midpoint does somewhat better on this problem than RK4, but the importance of this method is that the error can be expressed as a power series in $h$ exhibiting only even powers:
$$ y_n - y(x+H) = \sum_{i=1}^\infty \alpha_i h^{2i} $$
This explains the superior performance over RK4, but more importantly, this is the same property as 
we encountered in the trapezoidal rule. Does this sound familar?

In the same way we combined Richardson extrapolation with the trapezoidal rule to lead to Romberg intergration, we can apply Richardson extrapolation to the modified midpoint method, using succssive terms to cancel first the second-order error, then the fourth, and so on. The resulting method was derived by Burlisch and Stoer. For a sequence of sub-steps $h$ dividing $H$,
we can create a tableau just as in Romberg integration. For $y_{n,k}$ the result of taking $k$ subdivisions
to get to $y(x_n+H)$ by taking $h=H/n$, and starting the extrapolation by setting $T_{k,0} = y_{n,k}$,
we have the recursion
$$ T_{k,i+1} = T_{k,i} + \frac{T_{k,i} - T_{k,i-1}}{(n_k/n_{k-1})^2-1} $$
Now the elements of $T_{ki}$ are vectors of dependent variables, and the error for extrapolation $k$ can be defined as
$$ \epsilon_k = | T_{k,k} - T_{k,k-1} | = \left[\frac{1}{M}\sum_{i=1}^M (T_{k,k,i} - T_{k,k-1,i})^2\right]^{1/2} $$ 
Because we expect $\epsilon_k \sim H^{2k+1}$, we can use the same adaptive stepsize ideas we used with embedded Rungke-Kutta formulae to keep the error per step $H$ roughly constant.

In the following simple implementation, we use a sequence of subintervals defined by $n = (2,4,6,8,\dots)$ and use polynomial extrapolation on the sequence.

This is just an illustrative, example implementation. An industrial-stength version would provide options for dense output and absolute and relative error tolerances as well, and perhaps even provide for events as in the previous `scipy` example.

```python
class Info:
    """
    Simple structure class so we can return a structure of solution information
    from BulirschStoer, below
    """
    def __init__(self):
        pass

class BulirschStoer():
    def __init__(self, RHS, tol, yscale, *P):
        """
        Simple version of Bulirsch-Stoer ODE integrator
        (better off using odeint)
        
        RHS(t, y, *P) is RHS of ODE system
        yscale is scale factor per dimension for error tolerance
        tol is absolute error tolerance of scaled solution
        P are any extra parameters for RHS
        
        (adapted from Matlab code at github.com/MatthewPeterKelly/Bulirsch-Stoer/
        Internal arrays are still offset-1 (matlab style) not offset-0 (Python style)
        """
        self.P = P
        self.RHS = RHS
        self.tol = tol
        self.yscale = yscale

    def integrate(self, t, y0):
        """
        Wrapper to convert from offset-1 to offset-0 arrays
        """
        y, info = self.integrateInternal(t, y0)
        return y[:,1:].T, info
        
    def integrateInternal(self, t, y0):
        """
        Main recursive Bulirsch-Stoer routine;
        returns solution at requested values of independent variable t
        """
        
        nStepRefine = 3  # number of steps to replace a failed step
        nt = len(t)
        ny = len(y0)
        y = np.zeros( (ny, nt+1) )
        y[:,1] = y0
        
        # prepare structure to return information on solution
        info = Info()
        info.error = np.zeros_like(y)
        info.nFunEval = np.zeros(nt  +1, dtype=np.int32)
        
        # loop over time
        for i in range(2,nt+1):
            
            # take a BS step on the requested interval
            tSpan = (t[i-1  -1],t[i  -1])
            yF, stepInfo = self.integrateStep(tSpan, y[:,i-1])
            
            # if converged, keep the result
            if stepInfo.exit == 'converged':
                y[:,i] = yF
                info.error[:,i] = stepInfo.error
                info.nFunEval[i] = stepInfo.nFunEval
                
            # otherwise, recurse with smaller stepsize
            else: 
                time = np.linspace(tSpan[0], tSpan[1], self.nStepRefine+1)
                yTmp, infoTmp = self.integrateInternal(time, y[:,i-1])
                y[:,i] = zTmp[:,-1]
                info.error[:,i] = infoTmp.error[:,-1]
                info.nFunEval[i] = np.sum(infoTmp.nFunEval)

        return y, info
            
    def integrateStep(self, tSpan, y0):
        """
        Single step of Bulirsch-Stoer method
        """
        nRefineMax = 14
        
        n = 2*np.arange(1,nRefineMax+1)  # list of step refinements to take
                                         # in modified midpoint method
        
        ny = len(y0)
        if len(self.tol)==1:
            self.tol = self.tol*np.ones(ny)
            
        T = np.zeros( (ny, nRefineMax+1, nRefineMax+1) )  # extrapolated solution
        E = np.zeros( (ny, nRefineMax+1) )                # error
        
        info = Info()
        info.exit = 'maxRefine' # assume failure!
        
        for j in range(1, nRefineMax+1):
            
            y = self.modifiedMidpoint(tSpan, y0, n[j]) # estimate w/ current stepsize
            
            T[:,j,1] = y[:,-1]
            
            if j > 1:
                
                # polynomial extrapolation
                for k in range(2,j+1):
                    num = T[:,j,k-1] - T[:,j-1,k-1]
                    den = (n[j]/n[j-k+1])**2-1
                    T[:,j,k] = T[:,j,k-1] + num/den
                    
                # error
                E[:,j] = np.fabs(T[:,j,j-1] - T[:,j,j])
                
                # require convergence to tol in scaled solution 
                if np.all(E[:,j] < self.tol * self.yscale):
                    info.exit = 'converged'
                    break

        # provide information to calling routine
        info.error = E[:,j]
        info.nFunEval = np.sum(n[0:j])
        
        info.nRefine = j
        
        yF = T[:,j,j]
        return yF, info
    
    def modifiedMidpoint(self, tSpan, y0, n):
        """
        modified midpoint method
        """
        nt = n + 1
        ny = len(y0)
        
        t0 = tSpan[0]
        tF = tSpan[1]
        h = (tF-t0)/n
        t = np.zeros(nt+1)
        t[1:] = np.linspace(t0, tF, nt)
        y = np.zeros( (ny, nt+1) )
        
        y[:,1] = y0
        y[:,2] = y0 + h*self.RHS(t0, y0, *self.P)
        for i in range(3,nt+1):
            y[:,i] = y[:,i-2] + 2*h*self.RHS(t[i-1], y[:,i-1], *self.P)

        y[:,nt] = 0.5* ( y[:,nt] + y[:,nt-1] + h*self.RHS(t[nt], y[:,nt], *self.P) )
        
        return y
```
