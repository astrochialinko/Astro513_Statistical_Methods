```python
%matplotlib inline

import numpy as np
from scipy.fft import fftn, ifftn, fftshift, ifftshift
from scipy.special import erf, expi, exp1
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits import axes_grid1

def add_colorbar(im, ax, aspect=20, pad_fraction=0.5, **kwargs):
    """Add a vertical color bar to an image plot."""
    divider = axes_grid1.make_axes_locatable(im.axes)
    width = axes_grid1.axes_size.AxesY(im.axes, aspect=1./aspect)
    pad = axes_grid1.axes_size.Fraction(pad_fraction, width)
    cax = divider.append_axes("right", size=width, pad=pad)
    plt.sca(ax)
    return im.axes.figure.colorbar(im, cax=cax, **kwargs)

pi = np.pi

plt.rcParams['figure.dpi'] = 200
```

Here we present the solution of Poisson's equation
$$ \nabla^2 \phi = -\rho $$
with "open" (not periodic) boundary conditions(sometimes called "vacuum" boundaries).

As a test problem, we can use a Gaussian distribution of charge
$$ \rho = \frac{1}{(2\pi)^{3/2}\sigma^3} \exp\left(-\frac{x^2+y^2+z^2}{2\sigma^2}\right) $$
and solve the problem on the unit box $[-0.5,0.5]^3$ with $\sigma=0.05$.

The exact solution is 
$$ \phi(x,y,z) = \frac{1}{4\pi\sqrt{x^2+y^2+z^2}} \textrm{erf}\left(\frac{\sqrt{x^2+y^2+z^2}}{\sqrt{2}\sigma}\right) $$

```python
def rho3d(x,y,z,sigma):
    r = np.sqrt(x**2+y**2+z**2)
    return 1/((2*np.pi)**1.5 * sigma**3) * np.exp(-r**2/(2*sigma**2))

def phi3dExact(x,y,z,sigma):
    r = np.sqrt(x**2+y**2+z**2)
    # need l'Hopital's rule...
    return np.where(r==0, 1/( (2*pi)**1.5 * sigma ), \
        1/(4*np.pi*np.sqrt(x**2+y**2+z**2)) * erf(np.sqrt(x**2+y**2+z**2)/(np.sqrt(2)*sigma)) )
```

We'll sove this by convolving the Greens' function 
$$ G(\mathbf{r}) = -\frac{1}{4\pi | \mathbf{r} | }$$
with the density $\rho$. 

The Fourier transform of $G$ is
$$ \hat{G}(\mathbf{k}) = -\frac{1}{|\mathbf{k}|^2} $$
with discrete wave vectors in a box with $N$ cells per dimension
$$ k_x = \frac{2\pi n_x}{L}, k_y = \frac{2\pi n_y}{L}, k_z = \frac{2\pi n_z}{L} $$
where $L = N \Delta x$ is the length of the side of the box, $L=1$ in our case.

The Fourier Green's function has a singularity at $\mathbf{k} = 0$; some method of  "regularizing" it
must be employed. We will choose $\hat{G}(\mathbf{0}) = -1$

Our solution is not a periodic function. If we choose to compute the Fourier transforms twice as long in each direction ($2N$) and fill this extra space with zeros, there will be no power aliased into the problem space.
This is a somewhat brute-force approach, but effective; it is known as *zero-padding*.

If we wished to obtain a solution under periodic boundary conditions (as in cosmological calculations), this zero-padding would not be necessary as the Fourier transform implies a periodic function.

***

We can obtain a discrete Green's function as follows. A second-order inite-difference approximation to
the 1D Poisson problem is
$$ \frac{\phi_{j+1} - 2\phi_j + \phi_{j-1}}{h^2} = -\rho_j $$
The inverse DFT of $\hat{\phi}$ is 
$$ \phi_j = \frac{1}{Nh} \sum_{n=0}^{N-1} e^{-2\pi i n j/N} \hat{\phi}_n $$
Substituting this into the finite difference equation
$$ \frac{1}{N h^3} \left[ \sum_{n=0}^{N-1} e^{-2\pi i n (j+1)/N} \hat{\phi}_n - 2 \sum_{n=0}^{N-1} e^{-2\pi i n (j)/N} \hat{\phi}_n + \sum_{n=0}^{N-1} e^{-2\pi i n (j-1)/N} \hat{\phi}_n \right] = - \frac{1}{N h} \sum_{n=0}^{N-1} e^{-2\pi i n (j)/N} \hat{\rho}_n $$
Cancelling terms, this is
$$ \frac{1}{h^2} \left[ \sum_{n=0}^{N-1} e^{-2\pi i n/N} \hat{\phi}_n - 2 \sum_{n=0}^{N-1} \hat{\phi}_n + \sum_{n=0}^{N-1} e^{+2\pi i n /N} \hat{\phi}_n \right] = - \sum_{n=0}^{N-1} \hat{\rho}_n $$
This can only be true in general if 
$$  \frac{1}{h^2} \hat{\phi}_n \left[ e^{-2\pi i n/N} - 2 +  e^{+2\pi i n /N} \right] = - \hat{\rho}_n $$
which, using Euler's formula, is
$$  \frac{1}{h^2} \hat{\phi}_n \left[ 2 \cos(2\pi n/N) - 2 \right] = - \hat{\rho}_n $$
Using the half-angle identity $ 1-\cos(x) = 2\sin^2(x/2)$
we have
$$ \frac{1}{h^2} \hat{\phi}_n \left[ -4 \sin^2(\pi n/N) \right] = - \hat{\rho}_n $$
rearranging
$$ \hat{\phi}_n = \left(\frac{\sin(\pi n/N)}{h/2}\right)^{-2} \hat{\rho}_n $$
In three dimensions, then, we have
$$ G(\mathbf{k}) = \left[\left(\frac{\sin(k_x)}{h/2}\right)^2 + \left(\frac{\sin(k_y)}{h/2}\right)^2 + \left(\frac{\sin(k_z)}{h/2}\right)^2\right]^{-1} $$
with the wave vectors
$$ k_n = \frac{\pi n}{N}, \quad n=0,\dots,N-1 $$


***

Let's use a box size of $N=64$ on a side. The parameters of the problem are:

```python
N = 64
sigma = 0.05
xmin, xmax = np.array([-1,1]) * 0.5
L = xmax-xmin
```

We define a grid at the midpoints of the cells of rho, and assign it the source distribution $\rho$

```python
dx = (xmax-xmin)/N
x1d = np.arange(xmin+dx/2, xmax+dx/2, dx)
x, y, z = np.meshgrid(x1d,x1d,x1d)
rho = rho3d(x,y,z, sigma)
```

The wave vectors in our double-sized box are

```python
M = 2*N
k = 2*np.pi/M * np.indices((M,M,M))
```

The Green's function on the box is then

```python
Ghat = -1/( ((np.sin(0.5*k)/(0.5*dx))**2).sum(axis=0) + 1.0e-38)
Ghat[0,0,0] = -1
print(Ghat.shape)
```

We have to zero-pad the density:

```python
rhopad = np.pad(rho,(0,N))
print(rhopad.shape)
```

We now perform the convolution, and take the first octant of data

```python
rhohat = fftn(rhopad)
phihat = - rhohat * Ghat
phi = ifftn( phihat )[0:N,0:N,0:N].real
```

Plotting the errors, we see that there is a quite large error, but that it is nearly constant. Since we are most often interested in $\nabla \phi$, this constant offset can be ignored.

```python
ans = phi3dExact(x,y,z, sigma)
error = np.abs(ans-phi)/np.abs(ans)

fig, ax = plt.subplots(1,3)
c0 = ax[0].imshow(ans[:,:,N//2], extent=(-1,1,-1,1), origin='lower')
add_colorbar(c0, ax[0])
c1 = ax[1].imshow(phi[:,:,N//2], extent=(-1,1,-1,1), origin='lower')
add_colorbar(c1, ax[1])
c2 = ax[2].imshow(error[:,:,N//2], extent=(-1,1,-1,1), origin='lower')
add_colorbar(c2, ax[2])
plt.tight_layout()
```

Vico, Greengard, \& Ferrando (2016) (J. Comp. Phys. **323**, 191) introduce a more sophisticated approach which they comment is "both elementary and quite general, but seems to have been overlooked in the numerical analysis literature." As you will see, this is quite the understatement.

They start from the definition of the Fourier transform
$$ \hat{\phi}(\mathbf{s}) = \mathcal{F}[\phi(\mathbf{x})] = \int_D e^{-i\mathbf{s}\cdot\mathbf{x}} \phi(\mathbf{x}) d\mathbf{x} $$
where $D$ is the domain of integration, which they take to be our unit box, $D = [-0.5,0.5]^3$.

They note that, instead of the usual Green's function
$$ G(\mathbf{r}-\mathbf{r'}) = \frac{-1}{4\pi|\mathbf{r}-\mathbf{r'}|^2} $$
valid over all space, in a unit box the furthest distance between any two points is $\sqrt{3}$.  
They therefore write a truncated real-space Green's function as
$$ g_k^L(\mathbf{r}) = \frac{1}{4\pi}\frac{e^{-ikr}}{r}\mathrm{rect}\left(\frac{r}{2L}\right) $$
where the indicator function $\textrm{rect}(x)$ on the domain is
$$ \mathrm{rect}(x) = \left\{ \begin{array}{ll} 1, & |x|<\frac{1}{2}\\0, & |x| > \frac{1}{2} \end{array}\right. $$
and where $L>\sqrt{3}$. For the problem confined to the unit box, this is identical to the usual Green's function.

With this, the Poisson equation in 3D has the solution
$$ \phi(\mathbf{x}) = \frac{2}{(2\pi)^3} \int_{\mathbb{R}^3} e^{-i\mathbf{s}\cdot\mathbf{x}} \left(\frac{\sin(L|\mathbf{s}|/2)}{|s|}\right)^2 \hat{\rho}(\mathbf{s}) d\mathbf{s}$$
Since $\lim_{s\rightarrow0} \sin(s)/s = 0$, this Green's function is no longer singular, and there is no
problem of regularization.

The integral in this expression is sufficiently well-sampled if the mesh of the grid is four times finer than that used for the sampled $\rho$; twice for the zero-padding because this is a non-periodic solution, and
twice more because of the oscillatory behavior of the kernel. Thus, they employ a wave vector grid four times as fine (i.e. $4N$ per dimension if the solution is desired on a grid of $N$ per dimension) to compute a discrete representation of the Green's function, and then take a subset of $2N$ per side to do the padded convolution with $\rho$.

Here is an implementation of their method. For more details about the derivation, and for improved Green's functions for other elliptic operators, see their paper.

```python
xmin, xmax = np.array([-1,1])*0.5
L = xmax - xmin
pi = np.pi
N = 64
h = (xmax-xmin)/N
x1d = np.arange(xmin+h/2, xmax+h/2, h)
x, y, z = np.meshgrid(x1d,x1d,x1d)
```

```python
rho = rho3d(x,y,z,sigma)
```

Here are the wave vectors on the 4 times finer grid

```python
hs = 1/2
w1d = pi*np.arange(-N,N,hs)
w1d[2*N] = 1e-20
w1, w2, w3 = np.meshgrid(w1d, w1d, w1d)
```

and the truncated Green's function

```python
s = np.sqrt(w1**2 + w2**2 + w3**2)
LL = np.sqrt(3) 
Ghat = 2*(np.sin(LL*s/2.0)/s)**2 
```

We transform into real space and multiply the transformed function by $e^{-i \mathbf{k}\cdot\mathbf{r}}$

```python
xx1d = np.arange(4*xmin, 4*xmax, h)
xx, yy, zz = np.meshgrid(xx1d, xx1d, xx1d)
constant = np.exp(-1j*(xx+yy+zz)*N*pi) 
```

and then transform back to Fourier space. We take the central $2N\times2N\times2N$ values off the grid.
These become the final Green's function for use in the convolution with $\rho$

```python
T1 = ifftshift(ifftn(Ghat)) * constant
T = T1[N:3*N,N:3*N,N:3*N]
```

We still need to pad $\rho$ to prevent aliasing

```python
phi = ifftn(fftn(T) * fftn(np.pad(rho,(0,N))))[N:2*N,N:2*N,N:2*N].real
```

Compute the relative error in the solution; quite an improvement!
Now we have $T$ on the $2N$ grid, we can use it as often as we like. The computations
on the $4N$ grid are just to compute the operator and do not depend on $\rho$.

```python
ans = phi3dExact(x,y,z, sigma)
error = np.abs(ans-phi)/np.abs(ans)
L2error = np.sqrt(np.average(error**2))
print(f"L2 error: {L2error}")
fig, ax = plt.subplots(1,3)
c0 = ax[0].imshow(ans[:,:,N//2], extent=(-1,1,-1,1), origin='lower')
add_colorbar(c0, ax[0])
c1 = ax[1].imshow(phi[:,:,N//2], extent=(-1,1,-1,1), origin='lower')
add_colorbar(c1, ax[1])
c2 = ax[2].imshow(error[:,:,N//2], extent=(-1,1,-1,1), origin='lower')
add_colorbar(c2, ax[2])
plt.tight_layout()
```
