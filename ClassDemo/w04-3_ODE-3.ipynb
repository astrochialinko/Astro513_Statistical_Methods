# Scipy's solve_ivp driver and DOP853 method

A pretty good "workhorse" general purpose, adaptive timestep ODE solver which implements "events"

```python
%matplotlib inline

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 10/1.618]
plt.rcParams.update({'font.size': 18})       # make the fonts somewhat larger
```

We will start by finding the roots of the polynomial 
$$ p(x) = (x+6)(x-2)(x+2) $$
in two different ways.

We implement this polynomial function using Horner's method (minimum operation count) as:

```python
def func(x):
    # (x+6)*(x-2)*(x+2)
    # x**3 + 6*x**2 - 4*x - 24
    return -24 + x*(-4 + x*(6 + x))
```

The function looks like

```python
fig,ax = plt.subplots()
x = np.linspace(-8,4,100)
ax.plot(x, func(x))
ax.plot(x,np.zeros_like(x))
```

The first method is to use the `roots` function from Numpy which, given the coefficients of the polynomial (in order of *decreasing* monomial power), returns the roots (computed as the eigenvalues of the companion matrix):

```python
np.roots([1,6,-4,-24])
```

For our second method, we'll integrate the derivative of $p(x)$,
$$ p'(x) = 3x^2 + 12x - 4 $$
with an appropriate initial condition $p(x0)$. It is pretty clear from the plot that the roots all lie in the interval $[-8,4]$.

We will do so using Scipy's implementation of a driver similar to the one from NR in the previous examples, `solve_ivp`, and Scipy's implementation of the Dormand-Prince 8-5-3 stepper, `DOP853`.

We give `solve_ivp` the same RHS function we have been using. We can also give it an "events" function. When such a function is specified, the integrator will determine the point at which the function returns zero.
The numerical scheme used in DOP853 produces a 7-th order polynomial representing the solution on each interval. When the events function passes through zero on an interval (during a step), the appropriate root of the polynomial representation of the solution is obtained and returned in an additional array.


By giving an events function which simply returns the function value, we find the zero-crossings (roots) of the function.

Note that we are using a sequence of 7-th order polynomials over small intervals $h$ to approximate a single third-order polynomial -- extreme overkill for the given example!

```python
def derivs(x, y):
    """
    derivative of x**3 + 6*x**2 - 4*x - 24 in the form used by ODE integrators
    """
    return -4 + x*(12 + 3*x)

def event(x,y):
    """
    Zero-crossing event function
    """
    return y
```

We'll use this as our initial condition:

```python
func(-8)
```

Now integrate using solve_ivp with the DOP853 method, handing the event function, above, to the events parameter. When an events function is given, the integrator will return those values of the independent variable where `event(x,y)` becomes zero. In our case, therefore, we just have `event(x,y)` return `y`.

```python
result = solve_ivp(derivs, (-8,4), np.array([func(-8)]), max_step=0.5, method='DOP853', events=event)
```

The data stucture returned will contain an array of the times of the events and the value of the function at those times:

```python
print(f"roots =  {result.t_events[0].tolist()}")
print(f"p(roots) =  {result.y_events[0].tolist()}")
```

This event mechanism can be refined by giving the direction of the zero-crossing. To demonstrate this, we use the van der Pol oscillator example from before, but adding a driving term so that the system exhibits more complex behavior.

```python
class vanderPol():
    def __init__(self, mu, A, omega):
        self.mu = mu
        self.A = A
        self.omega = omega
        
    def derivs(self, t, y):
        """
        forced van del Pol oscillator 
        """
    
        v = y[1]
        x = y[0]
    
        return np.array([
            v,
            self.mu*(1-x**2)*v - x + self.A*np.sin(self.omega*t)     
                                     #    ^ driving term
        ])


def vdPevent(t, y):
    return y[1]  # look for where the derivative is zero
```

We've defined an event function which returns the derivative $v(t)$, so we are looking for where the derivative changes sign.

We also define an *attribute* of the `event` function which is $>0$; this means that `solve_ivp` will only give us the events where the derivative changes sign from negative to positive (moving upward across zero in the phase plot). The red dots at these crossings in the phase plot are thus a Poincare section.


```python
s = vanderPol(5,5,2.457)
warmupTime = 100
runTime = warmupTime + 100

vdPevent.direction = 1.0

y= solve_ivp(s.derivs, (0,warmupTime), np.array([1,-1]), atol=1e-8, rtol=1e-10, method='DOP853')
y0 = y.y[:,-1]
y= solve_ivp(s.derivs, (warmupTime, runTime), y0, atol=1e-8, rtol=1e-10, method='DOP853', 
             events=vdPevent, max_step=1)

fig, ax = plt.subplots(1,2, figsize=(15,6))
ax[0].plot(y.t, y.y[0,:],'r--')
ax[0].plot(y.t, y.y[1,:])
ax[0].plot(y.t_events,np.zeros_like(y.t_events),'ro')
ax[0].set_xlabel("$t$")
ax[0].set_ylabel("$dx/dt$")

ax[1].plot(y.y[0],y.y[1])
ax[1].plot(y.y_events[0][:,0],y.y_events[0][:,1],'ro')
ax[1].set_xlabel("$x(t)$")
ax[1].set_ylabel("$dx/dt$");
```

Finally, we can repeat our Lane-Emden calculations simply by looking for the first negative-going zero of the Lane-Emden function

```python
class LaneEmden():
    def __init__(self,n):
        self.n = n

    def derivs(self, x, y):
    
        dthetadx = y[1]
        dvdx = -2/x*y[1] - y[0]**self.n
    
        return np.array([ dthetadx, dvdx ])
```

```python
def event(t, y):
    return y[0]

event.direction = -1.0
event.terminal = True

n = 3
s = LaneEmden(n)

xi0 = 1e-6                            # starting value of xi for our numerical integration
theta0 = 1 - xi0**2/6 + n*xi0**4/120  # Taylor series solution to the DE near zero derived above
theta0p = -xi0/3 + n*xi0**3/30
y0 = np.array([ theta0,  theta0p])    # set IC's for numerical integration
print(f"IC at {xi0:10.5e}: {y0[0]:10.5e}, {y0[1]:10.5e}\n")

tmax = 100

y= solve_ivp(s.derivs, (xi0, tmax), y0, atol=1e-8, rtol=1e-10, t_eval=np.linspace(xi0,tmax,1000), method='DOP853', events=event, max_step=1)

print(f"xi1: {y.t_events[0]}")
print(f"theta'(xi1): {y.y_events[0][0,1]}\n")

fig, ax = plt.subplots()
ax.plot(y.t, y.y[0,:])
print(f"number of pts reported: {len(y.y[0,:])}")

ax.plot([0,8],[0,0],'k')
ax.plot(y.t_events,y.y_events[0][0,0],'ro')

plt.tight_layout()
plt.show()
```

Just for fun, let's compare the results of using Scipy's solve_ivp and DOP853 and my translation of from C++ to Python of Numerical Recipes' implementation of the same algorithm. We'll use the chaotic damped, driven pendulum as our example ODE system.

```python
class DampedDrivenPendulum:
    def __init__(self, Q, omegaD, A):
        self.gamma = 1/Q
        self.omegaD = omegaD
        self.A = A

    # return the rhs of our system
    def derivs(self, t, y):
        """
        here:
           y[0] is theta
           y[1] is omega
        returns:
           [ thetadot, omegadot ]
        """
        return np.array([
            y[1],
            -np.sin(y[0]) - self.gamma*y[1] + self.A*np.cos(self.omegaD*t)
        ])
```

Load in the version translated from Numerical Recipes & set up the problem

```python
from odeint import Odeint, Output
from stepperdopr853 import StepperDopr853

omegaD = 2/3
A = 1.5
Q = 1.38
rhs = DampedDrivenPendulum(Q, omegaD, A)

drivePeriod = 2*np.pi/omegaD
```

...and run them both.

The two phase plots show similar strange attractors, but looking at the Poincare sections carefully you can see that they are not idenitcal.

Note the exponential divergence between the two results. Why is this happening? Is this due to a bug?

```python
atol = 1.0e-08
rtol = 1.0e-08
h1 = 0.01
hmin = 0.0

nperperiod = 100
nperiods = 100
npts = nperperiod * nperiods
nwarm = 100
tstart = nwarm*drivePeriod
tmax = tstart+nperiods*drivePeriod

y0 = np.array([0,1])

# do a warmup to let the system forget its initial conditions
out = Output()
sol = Odeint(0, tstart, y0, atol, rtol, h1, hmin, rhs.derivs, StepperDopr853, out)
y0 = sol.integrate()

# integrate the system from these initial conditions using the translated version
out = Output(npts)
sol = Odeint(tstart, tmax, y0, atol, rtol, h1, hmin, rhs.derivs, StepperDopr853, out)
sol.integrate()
t, y = out.result()

# do the same thing using Scipy's version
# asking solve_ivp to report the solution at the same times employed by the integration above
sol1 = solve_ivp(rhs.derivs, (tstart, tmax), y0, atol=atol, rtol=rtol, t_eval=t, method='DOP853')
y1 = sol1.y.T

#t1 = sol1.t
y[:,0] -= 2*np.pi
y1[:,0] -= 2*np.pi

fig, ax = plt.subplots(1,3,figsize=(15,7))
ax[0].plot(y[:,0]/np.pi,y[:,1],lw=0.2)
ax[0].plot(y[::nperperiod,0]/np.pi,y[::nperperiod,1],'r.')
ax[0].set_xlim(-1.2,1.2)
ax[0].set_ylim(-2.5, 2.5)
ax[0].set_title("phase diagram\nNR")

ax[1].plot(y1[:,0]/np.pi,y1[:,1],lw=0.2)
ax[1].plot(y1[::nperperiod,0]/np.pi,y1[::nperperiod,1],'r.')
ax[1].set_xlim(-1.2,1.2)
ax[1].set_ylim(-2.5, 2.5)
ax[1].set_title("phase diagram\nScipy")

ax[2].semilogy(t, np.abs(y[::,0]-y1[::,0]),lw=0.5)
ax[2].set_title("distance between\nsolutions");
ax[2].set_xlabel('t')
plt.tight_layout()
```
