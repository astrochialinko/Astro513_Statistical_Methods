### The Particle-in-Cell (PIC) Method

PIC is an algorithm for integrating the equations of motion for collections of particles. As an exercise, we will use it first to solve a simple small-N N-body problem and then use it to examine the growth of large-scale structure.

The equations of motion for a gravitational N-body problem are simply
$$ \mathbf{a}_i = -G\sum_{\substack{j=0\\j\ne i}}^{N-1}\frac{m_j(\mathbf{r}_i - \mathbf{r_j})}{|\mathbf{r}_i-\mathbf{r}_j|^3} $$
where $\mathbf{r}_i$ and $\mathbf{a}_i$ are the position and the acceleration of body $i$.

We will integrate the equations of motion for the N-body problem using the second-order symplectic Verlet algorithm. As a shorthand, we will write the entire set of $N$ coordinates at timestep $n$ as $\mathbf{r}^n$ and $\mathbf{v}^n$.
Starting from the initial conditions $\mathbf{r}^0, \mathbf{v}^0$, the kick-drift-kick
version is
\begin{align}
\mathbf{v}^{n+\frac{1}{2}} &= \mathbf{v}^n + \frac{h}{2} \mathbf{a}(\mathbf{r}^n) \\
\mathbf{r}^{n+1} &= \mathbf{r}^{n-1} + h \mathbf{v}^{n+\frac{1}{2}} \\
\mathbf{v}^n &= \mathbf{v}^{n+\frac{1}{2}} + \frac{h}{2} \mathbf{a}(\mathbf{r}^{n+1}) 
\end{align}

The top-level algorithm is then:
   * Kick the velocities to $\mathbf{v}^{n+\frac{1}{2}}$ using the acceleration $\mathbf{a}(\mathbf{r}^n)$
   * Drift the positions to $\mathbf{r}^{n+1}$ using $\mathbf{v}^{n+\frac{1}{2}}$ 
   * Compute the acceleration at the new positions $\mathbf{a}(\mathbf{r}^{n+1})$ 
   * Kick the velocities to $\mathbf{v}^{n+1}$ using the acceleration $\mathbf{a}(\mathbf{r}^{n+1})$
   
The PIC portion of the algorithm is the way in which we will compute the accelerations.
Were we to compute the accelerations directly using the formula above, the result would take $O(N^2)$ work: $(N-1)$ terms in each sum, and one sum for each of the $N$ particles.
 
Instead, the PIC algorithm to compute the acceleration at the particle positions is:
   * Spread the mass of the particles onto an $M\times M\times M$ grid. 
   * Solve the Poisson equation for the potential $\phi$ on the grid using FFT's.
   * Take the derivative of $\phi$ on the grid to obtain $\mathbf{a}$ on the grid.
   * Interpolate from the acceleration on the grid back to the position of the particles.

As we shall see below, the first task is $O(N)$; the second is $O((M\log M)^3)$; the third is $O(M^3)$; and the fourth is
$O(N)$. Thus, if we have about the same number of particles as grid cells, $N\sim M^3$, we have an
$O(N\log N)$ algorithm for the N-body problem.

Clearly, the finite resolution of the grid will lead to accelerations which are *softened* at length scales smaller than the grid length. As we shall discuss in class, for many large-N N-body problems, this can be an advantage.



***


Thus, in addition to the KDK time integration scheme, we will need four ingredients:  
1. A method of "spreading" mass from particle positions to cells to form a density grid.
2. A method of interpolating from the grid cells to particle positions (1. must be the inverse of 2.).
3. A method of solving for the potential using FFTs.
4. A method for taking the derivative of $\phi$ to obtain $\mathbf{a}$ on the grid.

Let's examine these ingredients further:

#### Cloud-in-Cell Mass Assignment and Interpolation

The simplest scheme would be to find the cell in the grid containing the particle's position, and simply add that particle's mass to the cell. Dividing by the cell's volume would then give us a density grid.

Since we will need a uniformly-spaced density grid in order to use FFTs, we need a simple way to determine in which cell a particle lies. Assume our particles all lie in a cube $L$ on a side and that there are $M$ cells per dimension. The cell containing particle $m$ has indices
$$ i,j,k = \left\lfloor (x,y,z)_m \frac{N}{L} \right\rfloor $$
where $\lfloor . \rfloor$ is the floor function, the the largest integer less than or equal to its argument
(`floor(x)` in Python, not to be confused with `int(x)` or `trunc(x)`, both of which round toward zero).
This is known in computer science as a *hash function*, a function which takes a quantity and generates an index into a list from it.

We can then simply loop over all of the particles and add each particle's mass to the appropriate cell, dividing the cell masses afterward by the cell volume to obtain a density. This is known as *Nearest Grid Point* assignment.

A somewhat more accurate assignment method is known as *Cloud-in-Cell*. Instead of assuming the particle is a point mass and adding all of its mass to the cell which contains it, we assume that the particle is a cube the same size as a cell, but centered at the particle's position. We then compute the overlap of this cube with the eight grid cells it intersects, and assign an appropriate fraction of its mass to each of them.

In 2D, this looks like



The lower-left corner of the offset box is the particle's position $\mathbf{r}_m$, and the indicies of lower-left grid cell are   
`i = floor(r[m,0]*M/L-0.5); j = floor(r[m,1]*M/L-0.5)`   
The indices of the upper-right cell are   
`ip = mod(i+1,M); jp = mod(j+1,M)`   
where the modulus function takes care of wrapping the particle's position in the periodic box.   
Writing   
`ux = r[m,0]*M/L -0.5 - i; uy = r[m,1]*M/L -0.5 - j`
Once the `u`'s are computed, we need to periodically "wrap" the `i` indices as well   
`i = int(i)%M; j = int(j)%M`   
The volume of the contribution to the grid cells are then   
`rho[ i, j] += (1-ux)*(1-uy)*mass[m]`   
`rho[ip, j] +=   ux  *(1-uy)*mass[m]`   
`rho[ i,jp] += (1-ux)*  uy  *mass[m]`   
`rho[ip,jp] +=   ux  *  uy  *mass[m]`   
After adding the contributions of all particles, to convert to a density
we must divide by the a cell's volume   
`rho[:,:,:] /= dx**2`   
to get a density in $L$ units.

Three dimensions are handled with an additional `k`,`kp`, and `uz`, and dividing by `dx**3`.

The interpolation algorithm must be the exact inverse of the previous algorithm for assignment. For each particle, using the same definitions as above,
we have in 2D   
`ax[m,0] = (                                     \
            gradphix[ i,  j] * (1-ux) * (1-uy) + \
            gradphix[ip,  j] *   ux   * (1-uy) + \
            gradphix[ i, jp] * (1-ux) *   uy   + \
            gradphix[ip, jp] *   ux   *   uy     \
           ) * dx**2
`   
and so on for the other components of acceleration, with a similar four more terms in 3D.

#### Solving Poisson's Equation

Now that we have machinery for producing a density grid, we need to solve Poisson's equation for the
potential
$$ \nabla^2 \phi = 4\pi G\rho $$
using FFTs to do the convolution with the Green's function $G$:
$$ \phi = \mathcal{F}^{-1}\left[ \mathcal{F}[G] \mathcal{F}[\rho] \right] $$

For our N-body code we won't want periodic boundary conditions, so we will need to pad our $\rho$ grid by a factor of two in each dimension and use an appropriate Green's function. 

See the notebook *EllipticFourierOpenBoundaries* for more detail.

#### Finding the Acceleration

Two methods for taking the derivative of the potential suggest themselves:

1. Since we are already computing $\hat{\phi} = \mathcal{F}[G] \mathcal{F}[\rho]$, we can compute $\nabla\phi$ by computing three invese FFTs, using the identity
$$ \partial \phi/\partial x = \mathcal{F}^{-1}\left[ 2\pi i k_x \hat{\phi}\right] $$
The advantage of this is that the derivative is computed with spectral accuracy. The disadvantage is that
we will need to compute three inverse FFTs.   


2. By computing only one inverse FFT we obtain $\phi$ on the grid. We can then use finite differences to
differentiate $\phi$. To second order, we have the centered difference
$$ a_i \approx \frac{\phi_{i+1}-\phi_{i+1}}{2h} $$
If we desire more accuracy, we can always use a higher-order difference. To $O(h^4)$, we have the centered difference formula
$$ a_i \approx \frac{-\phi_{i+2} + 8 \phi_{i+1} - 8 \phi_{i-1} + \phi_{i-2}}{12 h^4} $$
This option is usually sufficiently accurate and somewhat faster than option 1.

Finite differences on a peridioc box are easy; we simply have to make sure to take the indices modulo $N$. `numpy` makes this particularly easy:    

`
def derivative(f, h, d):
        """
        Fourth-order finite difference approximation to first derivative of f array
        on periodic grid along dimension d with cell width h
        """
        return (1./12*np.roll(f, 2, axis=d) -  2./3*np.roll(f, 1, axis=d) \
               + 2./3*np.roll(f,-1, axis=d) - 1./12*np.roll(f,-2, axis=d))/h**4
`   

`numpy.roll(a, shift, axis=d)` simply shifts the entire grid `a` periodically by `shift` along the desired axis `d`.   

For non-periodic problems, as long as the box is large enough that particles never approach within three cells from the boundary, we can still use this trick.
