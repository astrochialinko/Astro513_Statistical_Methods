## Numerical Solution of ODE's

```python
# prepare for plotting & import necessary modules

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [12, 12/1.618]  # make the plots somewhat larger
plt.rcParams.update({'font.size': 18})       # make the fonts somewhat larger
```

Assume we have a first-order differential equation which can be expressed in the form

$$ \frac{dy}{dt} = RHS(y,t) $$

We will solve this on constant-interval mesh of the independent variable $t$ defined by

$$ t_n = t_0 + n h $$


##### Forward-Euler method

This simply solves the first-order forward difference approximation to $dy/dt$

$$ \frac{y_{n+1}-y_n}{h} = RHS(y_n,t_n)$$

as

$$ y_{n+1} = y_n + h RHS(y_n,t_n) $$
where $h$ is the "timestep", $\Delta t$.

```python
def forwardEuler(t, y, dt, RHS, *P):
    """
    Implements a single step of the forward-Euler finite-difference scheme
    P is used to pass additional parameters to RHS
    """
    return y + dt*RHS(t, y, *P)
```

We now need some sort of framework which will take this function and do the integration for us. One simple version looks like this:

```python
def odeSolve(RHS, interval, y0, h, method, *P):
    """
    ODE driver with constant step-size
    
    RHS(t, y, *P) takes the the current independent variable t and the current dependent variable y
    P is used to pass additional parameters to RHS
    interval is a 2-tuple with the beginning and ending value of the independent variable
    y0 contains the initial condition at interval[0]
    method(t,yh,RHS,*P) returns the solution at t+h
    """
    
    t0, tmax = interval
    
    # make the t-mesh; guarantees we stop precisely at tmax
    t = np.arange(t0,tmax+h,h)
    # allocate space for the solution
    y = np.zeros_like(t)
    # set the initial condition
    y[0] = y0
    
    # find out the size of the t-mesh, and then integrate forward one meshpoint per iteration of the loop
    n, = t.shape
    for i in range(0,n-1):
        y[i+1] = method(t[i], y[i], h, RHS, *P)
        
    # return the solution
    return t,y
```

Armed with this machinery, let's set up a simple problem and try it out.

We will solve the simple ODE

$$ \frac{dy}{dt} = - c y, \quad y(0) = 1 $$
with the solution
$$ y(t) = e^{-c t} $$

First, we provide a function to implement the RHS:

```python
def expRHS(t, y, c):
    """
    Implements the RHS (y'(x)) of the DE
    """
    return -c*y
```

Now we set up the problem to compute and plot the result, along with a plot of the magnitude of the errors 

```python
# set up problem
c = 1.0
h = 0.5
t0 = 0.0
y0 = 1.0
tmax = 5.0

# call the solver
t, y = odeSolve(expRHS, (t0, tmax), y0, h,forwardEuler, c)

# plot the result
fig,ax = plt.subplots(1,2)
ax[0].plot(t,y,'o', label='forward Euler')
tt = np.linspace(t0, tmax, 1000)
ans = np.exp(-c*tt)
ax[0].plot(tt,ans,'r', label='exact')
ax[0].set_xlabel('t')
ax[0].set_ylabel('y')

# along with the errors
ans = np.exp(-c*t)
err = ((ans-y))
ax[1].plot(t, err, 'r')
ax[1].plot(t, err, 'o')
ax[1].set_xlabel('t')
ax[1].set_ylabel('error')

ax[0].legend()

# this gives better spacing between axes
plt.tight_layout()
plt.show()
```

##### Explicit midpoint method

Next, we implement the explicit midpoint method, which is second-order in $h$

\begin{align}
\begin{split}
y_{n+\frac{1}{2}} &= y_n + \frac{h}{2} RHS(t_n, y_n) \\
y_{n+1} &= y_n + h RHS(t_n + \frac{1}{2}h, y_{n+\frac{1}{2}})
\end{split}
\end{align}

```python
def midpoint(t, y, dt, RHS, *P):
    """
    Implements a single step of the second-order, explicit midpoint method
    """
    yhalf = y + 0.5*dt*RHS(t, y, *P)
    return y + dt * RHS(t + 0.5*dt, yhalf, *P)
```

Inserting this in our solver setup from above, we have

```python
# set up problem
c = 1.0
h = 0.5
t0 = 0.0
y0 = 1.0
tmax = 5.0
# call the solver
t, y = odeSolve(expRHS, (t0, tmax), y0, h, midpoint, c)  # only change from above!

fig,ax = plt.subplots(1,2)
ax[0].plot(t,y,'o')
tt = np.linspace(t0, tmax,1000)
ans = np.exp(-c*tt)
ax[0].plot(tt,ans,'r')
ax[0].set_xlabel('t')
ax[0].set_ylabel('y')

ans = np.exp(-c*t)
err = ans-y
ax[1].plot(t, err, 'r')
ax[1].plot(t, err, 'o')
ax[1].set_xlabel('t')
ax[1].set_ylabel('error')

plt.tight_layout()
plt.show()
```

##### RK4

Finally, we implement a version of the fourth-order Runge-Kutta method RK4:

```python
def RK4(t, y, h, RHS, *P):
    """
    Implements a single step of a fourth-order, explicit Runge-Kutta scheme
    """
    thalf = t + 0.5*h
    k1 = h * RHS(t, y, *P)
    k2 = h * RHS(thalf, y + 0.5*k1, *P)
    k3 = h * RHS(thalf, y + 0.5*k2, *P)
    k4 = h * RHS(t + h, y + k3, *P)
    return y + (k1 + 2*k2 + 2*k3 + k4)/6
```

```python
# set up problem
c = 1.0
h = 0.5
t0 = 0.0
y0 = 1.0
tmax = 5.0
# call the solver
t, y = odeSolve(expRHS, (t0, tmax), y0, h, RK4, c)  # only change from above!

fig,ax = plt.subplots(1,2)
ax[0].plot(t,y,'o')

ans = np.exp(-c*tt)
ax[0].plot(tt,ans,'r')
ax[0].set_xlabel('t')
ax[0].set_ylabel('y')

ans = np.exp(-c*t)
err = np.abs((ans-y))
ax[1].plot(t, err, 'r')
ax[1].plot(t, err, 'o')
ax[1].set_xlabel('t')
ax[1].set_ylabel('|error|')

plt.tight_layout()
plt.show()
```

### Systems of First-Order ODEs

Next, we turn to systems of ODE's. We'll take as our example the Lotke-Volterra equations, a simple model of population dynamics in an ecosystem (with many other uses as well).

Imagine a population of rabbits and of foxes on a small island. The rabbits eat a plentiful supply of grass and
would breed like, well, rabbits, with their population increasing exponentially with time in the absence of preditors. The foxes eat the rabbits, and would die out exponentially in time with no food supply. The rate at which foxes eat rabbits depends upon the product of the fox and rabbit populations.

The equations for the population of the rabbits $R$ and foxes $F$ in this simple model is then

\begin{eqnarray*}
\frac{dR}{dt} &= \alpha R - \beta R F \\
\frac{dF}{dt} &= \delta R F - \gamma F
\end{eqnarray*}

Without the cross terms in $RF$, these are just two decay equations of the form we have used as an example above.

A random set of parameters (I am not a biologist!) might be that a rabbit lives two years, so $\alpha=1/2$ and
a fox lives 5 years, so $\gamma=1/5$. Let's pick the other parameters as $\beta = 1/80$ and $\delta = 1/200$.

We can express the unknown populations as a vector of length two: $y = (R, F)$. The rate of change of populations then can also be expressed as a vector $dy/dt = (dR/dt, dF/dt)$. With such a definition, we can write the RHS function of our system as

```python
def lvRHS(t, y, *P):
    # Lotke-Volterra system RHS
    
    # unpack the parameters from the array P
    alpha, beta, gamma, delta = P

    # make temporary variables with rabbit and fox populations
    R = y[0]
    F = y[1]
    
    # LV system
    dRdt = alpha * R     -  beta * R * F
    dFdt = delta * R * F - gamma *     F
    
    # return an array of derivatives with same order as input vector
    return np.array([ dRdt, dFdt ])
```

We now have to generalize our odeSolve function to allow for more than one equation

```python
def odeSolve(RHS, interval, y0, h, method, *P):
    """
    ODE driver with constant step-size, allowing systems of ODE's
    """
    
    t0, tmax = interval
    
    # make array of times and find length of array
    t = np.arange(t0,tmax+h,h)
    ntimes,  = t.shape
    
    # find out if we are solving a scalar ODE or a system of ODEs, and allocate space accordingly
    
    if type(y0) in [int, float]:  # check if primitive type -- means only one eqn
        neqn = 1
        y = np.zeros( ntimes )
    else:                         # otherwise assume a numpy array -- a system of more than one eqn
        neqn, = y0.shape
        y = np.zeros( (ntimes, neqn) )

    # set first element of solution to initial conditions (possibly a vector) 
    y[0] = y0

    # time marches on...
    for i in range(0,ntimes-1):
        y[i+1] = method(t[i], y[i], h, RHS, *P)

    return t,y
```

Now we can solve our system of two coupled ODEs. Note that the solution is now a vector of 2D vectors... the first index is the solution timestep number, the second the variable (0 for rabbits, 1 for foxes)

```python
alpha = 1/2
beta = 1/80
gamma = 1/4
delta = 1/200

h = 0.2
t0 = 0.0
y0 = np.array([ 30, 10 ])
tmax = 50

# call the solver
t, y = odeSolve(lvRHS, (t0, tmax), y0, h, RK4, alpha, beta, gamma, delta)

fig,ax = plt.subplots(1,2)
ax[0].plot(t,y[:,0],'b', label='prey')
ax[0].plot(t,y[:,1],'r', label='preditor')
ax[0].set_xlabel('time')
ax[0].set_ylabel('population')
ax[0].legend()

ax[1].plot(y[:,0],y[:,1],'g')
ax[1].set_xlabel('prey')
ax[1].set_ylabel('preditor')

plt.tight_layout()
plt.show()
```

### Higher-order derivatives in ODEs

The "trick" to solving ODEs with higher derivatives is turning them into systems of first-order ODEs.

As a simple example, consider the second-order differential equation describing the van der Pol oscillator

$$ \frac{d^2 x}{dt^2} - a (1-x^2) \frac{dx}{dt} + x = 0 $$

We turn this into a pair of first-order ODEs by defining an auxiliary function $v(t) = dx/dt$ and writing the system as

\begin{align}
\begin{split}
\frac{dx}{dt} &= v \\
\frac{dv}{dt} &= a (1-x^2) v - x\\
\end{split}
\end{align}

Note that there are only functions (and the independent variable) on the RHS; all derivatives are on the LHS.

Now that we have a system of first-order equations, we can proceed as above. A function describing
the RHS of this system is

```python
def vdpRHS(t, y, a):
    # we store our function as the array [x, x']
    return np.array([
        y[1],                          # dx/dt = v
        a*(1-y[0]**2)*y[1] - y[0]      # dv/dt = a*(1-x**2)*v - x
        
    ])
```

```python
a = 15 # parameter

h = 0.01
t0 = 0.0
y0 = np.array([ 0,  0.1])
tmax = 100

# call the solver
t, y = odeSolve(vdpRHS, (t0, tmax), y0, h, RK4, a)

fig,ax = plt.subplots(1,2)
ax[0].plot(t,y[:,0],'b', label='x')
ax[0].plot(t,y[:,1],'r--', label='v')
ax[0].set_xlabel(r'$t$')
ax[0].set_ylabel(r"$x(t), v(t)$")
ax[0].legend()
ax[0].set_title(f"van der Pol Oscillator for a={a}")

ax[1].plot(y[:,0],y[:,1])
ax[1].set_xlabel(r"$x(t)$")
ax[1].set_ylabel(r"$v(t)$")
   
plt.tight_layout()
plt.show()
```

A somewhat more complex example is the Lane-Emden equation, which is really just Poisson's equation in spherical symmetry for the graviational potential of a self-gravitating fluid whose pressure is related to its density as $P\propto\rho^\gamma$. Such a system is called a *polytrope*, and is often used in astrophysics as a simple model for the structure of systems such as stars in which outward pressure and inward gravity are in equilibrium.

Let $\xi$ be the dimensionless radius of the system, and let $\theta$ be related to the density as
$\rho = \rho_c \theta^n$, where $\rho_c$ is the density at the origin, and $n = 1/(\gamma-1)$ is known as the *polytropic index*. We then have the dimensionless second-order differential equation

$$ \frac{1}{\xi^2}\frac{d}{d\xi}\left(\xi^2\frac{d\theta}{d\xi}\right) + \theta^n = 0 $$

Note that the first term is just the divergence $\nabla\cdot\theta$ in spherical symmetry.

If we expand out the first term, we have 

$$ \frac{d^2\theta}{d\xi^2} + \frac{2}{\xi}\frac{d\theta}{d\xi} + \theta^n = 0 $$

Defining an auxiliary function $v(\xi) = d\theta/d\xi$, we can then convert this into a system of two first-order ODEs:

\begin{align}
\begin{split}
\frac{d\theta}{d\xi} &= v \\
\frac{dv}{d\xi} &= -\frac{2}{\xi} v - \theta^n \\
\end{split}
\end{align}

Again, we have derivatives only on the LHS and no derivatives on the RHS of our system.

First, have a look at the behavior of this system at the origin ($\xi=0$):  
*Analytically*, this is no problem at all: $v/\xi\rightarrow0$ as $\xi\rightarrow0$.  
*Numerically*, we have a problem: at $\xi=0$, a naive solution will divide by zero.

What to do?


The first approach is to take care of the problem in our RHS function:

```python
def leRHS(x, y, n):
    
    dThetadx = y[1]
    
    if x==0:
        dvdx = -y[0]**n
    else:
        dvdx = -2/x*y[1] - y[0]**n
    
    return np.array([ dThetadx, dvdx ])
```

It turns out that the solution $\theta(t)$ is quite smooth -- the higher derivatives are well-behaved -- and if our stepsize $h$ is reasonably small, this will give a perfectly good solution.

The approach is somewhat clunky, however.  First, you would have to convince yourself that in fact $v(\xi)\rightarrow0$ faster than $\xi$ (don't just take my word for it!). Second, inserting `if` statements into functions is often problematic (unless the function is really discontinuous!). 

Instead, we could use the simpler RHS function

```python
def leRHS(x, y, n):
    dThetadx = y[1]
    
    dvdx = -2/x*y[1] - y[0]**n
    
    return np.array([ dThetadx, dvdx ])
```

and expand the solution in a Taylor series about the origin to get a starting value for our numerical integration at a small distance away from the origin. (The result of doing so will convince you of the correctness of the first approach as well!) To do this, write 

$$\theta(\xi) = a_0 + a_1 \xi + a_2 \xi^2 + \dots$$

The first thing to notice is that, by the symmetry of a sphere, only even powers of $\xi$ will appear in the solution.
Thus we will have

$$ \theta(\xi) = a_0 + a_2 \xi^2 + a_4 \xi^4 + \dots$$

By the boundary condition $\theta(0) = 1$, we have immediately that $a_0 = 1$.

Next, substitute $\theta(\xi) = 1 + a_2 \xi^2 + a_4 \xi ^4 + O(\xi^6)$ into the Lane-Emden equation. $\theta$ and its first two derivatives are

\begin{align}
\begin{split}
\theta(\xi) &= 1 + a_2 \xi^2 + a_4 \xi^4 + O(\xi^6)\\
\theta'(\xi) &= 2 a_2 \xi + 4 a_4 \xi^3 + O(\xi^5) \\
\theta''(\xi) &= 2 a_2 + 12 a_4 \xi^2 + O(\xi^4)
\end{split}
\end{align}

Putting these into the Lane-Emden equation, we have
\begin{align}
\begin{split}
2 a_2 + 12 a_4 \xi^2 + O(\xi^4) + \frac{2}{\xi} (2 a_2 
\xi+ 4 a_4 \xi^3 + O(\xi^5))  &= -\theta^n \\
6 a_2 + 20 a_4 \xi^2 + O(\xi^4) &= -\theta^n
\end{split}
\end{align}

At the inner boundary $\theta(0)=1$, and thus we have $a_2 = -1/6$. Away from zero, then, we have

\begin{align}
\begin{split}
-1 + 20 a_4 \xi^2 + O(\xi^4) &= -\left(1 - 1/6 \xi^2 + a_4 \xi^4 + O(\xi^6)\right)^n
\end{split}
\end{align}

The RHS is $ 1 - n \xi^2/6 + O(\xi^4)$, and so we must have $a_4 = n/120$.

Thus, a series expansion of the solution about the origin is

$$ \theta(\xi) = 1 - \frac{1}{6}\xi^2 + \frac{n}{120} \xi^4 + O(\xi^6) $$

We can now use this expansion to take a first step slightly away from the origin before beginning our
numerical integration, thus avoiding the divide by zero. Note that this series solution near the origin is $O(h^6)$ and so is a good match for RK4 if we take the same (or smaller) step-size.

```python
# Polytropic index (corresponding to gamma=5/3, a perfect monatomic gas)
n = 3

xi0 = 0.01                            # starting value of xi for our numerical integration
theta0 = 1 - xi0**2/6 + n*xi0**4/120  # Taylor series solution to the DE near zero derived above
theta0p = -xi0/3 + n*xi0**3/30
y0 = np.array([ theta0,  theta0p])    # set IC's for numerical integration
print(f"IC at {xi0:10.5e}: {y0[0]:10.5e}, {y0[1]:10.5e}")

h = 0.1
ximax = 8

# call the solver
t, y = odeSolve(leRHS, (xi0, ximax), y0, h, RK4, n)

fig,ax = plt.subplots()
ax.plot(t,y[:,0],'b', label=r'$\theta(\xi)$')
ax.plot(t,y[:,1],'r--', label=r'$d\theta/d\xi$')
ax.plot([0,ximax],[0,0],'k')
ax.set_xlabel(r'$\xi$')
ax.set_title(f"Lane Emden Equation for n={n}")
ax.legend()

plt.tight_layout()
plt.show()
```

For values of $n\le5$, the solutions of the Lane Emden equation (the so-called Lane-Emden functions of index $n$) decrease to zero at finite $\xi$. Since this is the radius at which the density goes to zero, we can interpret it as the surface of the self-gravitating body (for example, the radius of the star). Knowing this value $\xi_1$ is thus interesting... Let us see how to determine it numerically.

Cleary, we are looking for the solution to $\theta(\xi_1)=0$; this is just root-finding, which we already know how to do. Instead of using some closed-form function, however, the value of the function $\theta(\xi)$ must in this case be determined numerically. But we have just figured out how to do this!

Let's use the bisection method for our root-finding algorithm; here is a quick and dirty version (no error checking!)

```python
def bisection(func, low, high, eps, *P):
    flow = func(low, *P)
    fhigh = func(high, *P)
    
    mid = 0.5*(low+high)
    fmid = func(mid,*P)
    
    while (high-low) > eps:
        if fmid*flow < 0:
            high = mid
            fhigh = fmid
        else:
            low = mid
            flow = mid
            
        mid = 0.5*(low+high)
        fmid = func(mid,*P)
            
    return low
```

Now let us make a function which returns $\theta(\xi)$, the solution to the Lane-Emden equation at $\xi$ determined by numerical integration

```python
def theta(xi, n):
    h = 1e-4
    xi0 = 1e-7
    theta0 = 1 - xi0**2/6 + n*xi0**4/120
    theta0p = -xi0/3 + n*xi0**3/30
    y0 = np.array([ theta0,  theta0p])
    t, y = odeSolve(leRHS, (xi0, xi), y0, h, RK4, n)

    # return the last value of theta
    return y[-1,0]
```

Using these functions, we can compute the surface radius of the polytrope. We can see from the plot above that the surface lies somewhere on the interval $[6,8]$

```python
n = 3
xi1 = bisection(theta, 6, 8, 1e-4, n)
print(f"xi_1 = {xi1:7.5f}")
```

A more careful treatment gives a value $\xi_1 = 6.89685...$, so we are doing pretty well...
