### Stiff Systems of ODEs

```python
# inline saves the plot in the notebook for later viewing w/o re-running
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
import scipy.linalg as linalg

plt.rcParams['figure.dpi'] = 200  # good way to scale up plots
```

Often in modeling a physical system, one runs into two or more characteristic timescales which differ greatly in magnitude. Consider the following example

\begin{align}
u' &= 998 u + 1998 v & u(0) = 1 \\
v' &= -999u - 1999v & v(0) = 0
\end{align}
The solution is
\begin{align}
u & = 2 e^{-x} - e^{-1000x} \\
u &= -e^{-x} + e^{-1000x}
\end{align}

Clearly, one would need a very small timestep for stability when $x$ is near zero as the evolution is dominated by the timescale $10^{-3}$. At larger $x$, the initial transient behavior at short timescales is no longer apparent in the solution, but the presence of the small timescale will still lead to a very tight constraint on the timestep for stability.

Let's try using our usual machinery for solving this system, starting with the explicit Runge-Kutta DOP853 integrator from last class:

```python
class stiffOne:
    def __init__(self):
        c00 = 998
        c01 = 1998
        c10 = -999
        c11 = -1999
        
        self.A = np.array([[c00,c01],[c10,c11]])
        self.y0 = np.array([1.0,0.0])
        
    def derivs(self, t, y):
        return self.A@y
    
    def exact(self, t):
        return np.array([2*np.exp(-t) - np.exp(-1000*t), -np.exp(-t) + np.exp(-1000*t)])
```

```python
s = stiffOne()
sol = solve_ivp(s.derivs, (0,5), s.y0, rtol=1e-6, atol=1e-6, method='DOP853')
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(sol.t, sol.y[0], label='u')
ax[0].plot(sol.t, sol.y[1], label='v')
ax[0].plot(sol.t, s.exact(sol.t)[0])
ax[0].plot(sol.t, s.exact(sol.t)[1])
ax[0].set_xlabel('t')
ax[0].set_ylabel('u,v')
ax[0].legend()

dt = sol.t[1:] - sol.t[:-1] 
ax[1].semilogy(sol.t[:-1], dt, '.');
ax[1].set_xlabel('t')
ax[1].set_ylabel(r'$\Delta t$')
plt.tight_layout();
```

```python
print(f"size of first timestep: {sol.t[1]-sol.t[0]: 0.2e}")
```

A *very* small timestep had to be employed at the beginning to resolve the inital transient! You certainly would want an adaptive timestep; otherwise, you would have to take $\sim10^4$ timesteps to do the problem.

But this is only the beginning of our problems...

Another example comes from a set of chemical rate equations arising in a combustion problem.  We have three chemical species with
the reactions
\begin{align}
S_1 &\rightarrow S_2  \\
S_2 + S_3 &\rightarrow S_1 + S_3 \\
S_2 + S_2 &\rightarrow S_3 + S_3
\end{align}
The evolution of the concentrations of these species is
\begin{align}
\frac{d S_1}{dt} &= -0.04 S_1 + 10^4 S_2 S_3\\
\frac{d S_2}{dt} &= \phantom{-}0.04 S_1 - 10^4 S_2 S_3 - 3\times 10^7 S_2^2\\
\frac{d S_3}{dt} &= 3\times 10^7 S_2^2\\
\end{align}
One can easily see that the steady-state solution is $t\rightarrow\infty$ is $S_1=S_2=0, S_3 =1$.
However, this steady state is achived only at $t\sim10^{11}$

This system has a range of timescales spanning amost $10^9$.

Let's try this one out, again using the explicit Runge-Kutta DOP853

```python
class reactSys:
    def __init__(self):
        self.y0 = np.array([1, 0, 0])
        
    def derivs(self,t,y):
        x, y, z = y
        return np.array([
            -0.04 * x + 1.e4 * y * z,
             0.04 * x - 1.e4 * y * z - 3.e7 * y**2,
             3.e7 * y**2
        ])
    
    def jacobian(self,t,y):
        x,y,z = y
        return np.array([
            [-0.04,  1.0e4*z,            1.0e4*y],
            [ 0.04, -1.0e4*z - 6.0e7*y, -1.0e4*y],
            [    0,            6.0e7*y,        0]
        ])
```

```python
s = reactSys()
solDOP = solve_ivp(s.derivs, (0,100), s.y0, method='DOP853')
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(solDOP.t, solDOP.y[0],'b',label=r'$S_1$')
ax[0].plot(solDOP.t, solDOP.y[1]*1e4, 'r', label=r'$10^4 S_2$')
ax[0].plot(solDOP.t, solDOP.y[2], 'g', label=r'$S_3$')
ax[0].legend(loc=1)
ax[0].set_xlabel('t')
ax[0].set_ylabel(r'$S_i$')

dt = solDOP.t[1:] - solDOP.t[:-1] 
ax[1].semilogy(solDOP.t[:-1], dt)
ax[1].set_xlabel('t')
ax[1].set_ylabel(r'$\Delta t$')

plt.tight_layout();
```

You can see for this system that the timestep chosen by DOP853 stays small for all time; indeed, it is decreasing even where the solution is becoming smoother!

Something funny is also going on with the solution for $S_1$; it appears to be oscillating; try expanding the x-axis:

```python
s = reactSys()
soltmp = solve_ivp(s.derivs, (0,0.5), s.y0, method='DOP853')

fig, ax = plt.subplots(2,1)
ax[0].plot(soltmp.t, soltmp.y[0],'b',label=r'$S_1$')
ax[0].plot(soltmp.t, soltmp.y[1]*1e4, 'r', label=r'$10^4 S_1$')
ax[0].plot(soltmp.t, soltmp.y[2], 'g', label=r'$S_3$')
ax[0].legend(loc=1)
ax[0].set_xlabel('t')
ax[0].set_ylabel(r'$S_i$')

dt = soltmp.t[1:] - soltmp.t[:-1] 
ax[1].semilogy(soltmp.t[:-1], dt)
ax[1].set_xlabel('t')
ax[1].set_ylabel(r'$\Delta t$')

plt.tight_layout();
```

Not terribly inspiring...

Ignoring the oscillations for the moment, this method seems wildly inefficient with such small timesteps. Poking around in the returned structure, we find an enormous number of timesteps giving an enormous number of evaluations of the RHS. Remember, we are only integrating to $t=100$; the reaction doesn't complete until $t\sim10^{11}$.

```python
print(solDOP.message)
print(f"number of steps: {solDOP.t.shape[0]}")
print(f"number of function evaluations: {solDOP.nfev}")
```

The extremely small timesteps despite the smoothness of the solution arises from the speed of the $S_2+S_2$ reaction relative to the $S_1\rightarrow S_2$ reaction. A tiny error in the solution at one time will lead to a large change in the solution at the next timestep.

Systems such as these are known as *stiff systems*. The solutions can be quite smooth, yet still require very small timesteps to resolve the short-timescale behavior. 

The cure for such pathology is to go to *implicit differencing*. For our system
$$ \frac{d\mathbf{y}}{d\mathbf{t}} = \mathbf{f}(t, \mathbf{y}) $$
we could try the first-order, implicit backward Euler scheme
$$ \mathbf{y}_{n+1} = \mathbf{y}_n + h \mathbf{f}(t_{n+1}, \mathbf{y}_{n+1}) $$
but now we have to do a non-linear solve to find $\mathbf{y}_{n+1}$ since our reaction system is a one of quadratic polynomials.

<br>

We will do this non-linear solve using the multidimensional version of the Newton-Raphson method. Recall
Newton's method for solving $f(y) = 0$ was, given an initial guess $y_0$, the iteration
$$ y_{i+1} = y_i - \frac{f(y_i)}{f'(y_i)} $$
In more than one dimension, this generalizes to 
$$ \mathbf{y}_{i+1} = \mathbf{y}_i - \mathbf{J}^{-1}\mathbf{y}_i $$
where $\mathbf{J}$ is the Jacobian matrix of partial derivatives
$$ J_{ij} = \frac{\partial f_i}{\partial y_j} $$

<br>

The first thing to try is to *linearize* the system: take the first two terms in the multidimensional Taylor series, approximating
$$ \mathbf{f}(t_{n+1}, \mathbf{y}_{n+1}) \approx \mathbf{f}(t_n, \mathbf{y}_n) + \mathbf{J}(t_n, \mathbf{y}_n)(\mathbf{y}_{n+1}-\mathbf{y}_n) $$

We then have
$$ \mathbf{y}_{n+1} = \mathbf{y}_n + h\left[ \mathbf{f}(t_n,\mathbf{y}_n) + \mathbf{J}_n (\mathbf{y}_{n+1}-\mathbf{y}_n)\right] $$
or, rearranging,
$$ \mathbf{y}_{n+1} = \mathbf{y}_n + h\left[\mathbf{I} - h J_n\right]^{-1} \mathbf{f}_n $$
If we take a small enough timestep $h$, the previous solution may be a good enough guess so we need only take one Newton iteration. Such a scheme is known as a *semi-implicit* scheme since we aren't really solving the full non-linear system exactly at each timestep. The method is more expensive per timestep than an explicit one of the same order since we must solve a linear system.

Let's try this out with a simple implementation

```python
def semiImplicitEuler(t, y, dt, RHS, Jac, *P):
    n = y.shape[0]
    #return y + dt * linalg.inv(np.eye(n) - dt*Jac(t,y,*P)) @ RHS(t,y,*P)
    return y + dt * linalg.solve((np.eye(n)-dt*Jac(t,y,*P)), RHS(t,y,*P))
```

```python
def odeSolve(RHS, Jac, interval, y0, h, method, *P):
    """
    ODE driver with constant step-size, allowing systems of ODE's
    """
    
    t0, tmax = interval
    
    # make array of times and find length of array
    t = np.arange(t0,tmax+h,h)
    ntimes,  = t.shape
    
    # find out if we are solving a scalar ODE or a system of ODEs, and allocate space accordingly
    
    if type(y0) in [int, float]:  # check if primitive type -- means only one eqn
        neqn = 1
        y = np.zeros( ntimes )
    else:                         # otherwise assume a numpy array -- a system of more than one eqn
        neqn, = y0.shape
        y = np.zeros( (ntimes, neqn) )

    # set first element of solution to initial conditions (possibly a vector) 
    y[0] = y0

    # time marches on...
    for i in range(0,ntimes-1):
        y[i+1] = method(t[i], y[i], h, RHS, Jac, *P)

    return t,y
```

Despite the tiny timestep chosen by DOP853, we will try this with a timestep of $h=10^{-2}$

```python
s = reactSys()
h = 1e-2
tE, yE = odeSolve(s.derivs, s.jacobian, (0,100), s.y0, h, semiImplicitEuler)
```

```python
def plotIt(t, y, tt=None, yy=None):
    fig, ax = plt.subplots(2,1)
    ax[0].plot(t, y[:,0],'b', label=r'$S_1$')
    ax[0].plot(t, y[:,1]*1e4, 'r', label=r'$10^4\ S_2$')
    ax[0].plot(t, y[:,2],'g', label=r'$S_3$')

    if tt is not None:
        ax[0].plot(tt, yy[:,0],'b.')
        ax[0].plot(tt, yy[:,1]*1e4,'r.')
        ax[0].plot(tt, yy[:,2],'g.')
        
    ax[0].set_ylim(-0.1,1.1)
    ax[0].set_xlabel('t')
    ax[0].set_ylabel(r'$S_i$')
    ax[0].legend(loc=1)

    ax[1].semilogx(t, y[:,0],'b')
    ax[1].semilogx(t, y[:,1]*1e4, 'r')
    ax[1].semilogx(t, y[:,2],'g')
        
    if tt is not None:
        ax[1].semilogx(tt, yy[:,0],'b.')
        ax[1].semilogx(tt, yy[:,1]*1e4,'r.')
        ax[1].semilogx(tt, yy[:,2],'g.')

    ax[1].set_xlabel('t')
    ax[1].set_ylabel(r'$S_i$')
    
    return ax
```

To have an "exact" solution with which to compare, we will use a 5$^{th}$-order adaptive Radau method which we will describe below; for now, we can assume it is designed to handle this kind of problem.

```python
# Use a method yet to be described to get the "correct answer"
sol = solve_ivp(s.derivs, (0,100), s.y0, rtol=1e-6, method='Radau')

ax = plotIt(tE, yE, sol.t, sol.y.T)
ax[0].set_title('Implicit Euler');
plt.tight_layout();
```

Remember that the Euler method is only first order in $h$, but becaue this method is implicit, there is no stability limit on the timestep. The early-time solution is quite inaccurate, but the solution recovers once the derivatives become smaller.

Despite a timestep much larger than the one initially used by the explicit Runge-Kutta `DOP853`, the correct long-term behavior of the solution is retained.

Let's go a step further and derive the fully-implicit form of RK2. One form, based on two-point Gaussian integration and fourth-order accurate, is
\begin{align}
\mathbf{k}_1 &= \mathbf{f}\left(t_n + c_1h, \mathbf{y}_n + h(a_{11}\mathbf{k}_1 + a_{12}\mathbf{k}_2)\right)\\
\mathbf{k}_2 &= \mathbf{f}\left(t_n+c_2h, \mathbf{y}_n + h(a_{21}\mathbf{k}_1 + a_{22}\mathbf{k}_2)\right)\\
\mathbf{y}_{n+1} &= \mathbf{y}_n + \frac{h}{2}(\mathbf{k}_1 + \mathbf{k}_2)
\end{align}
with constants
\begin{align}
a_{11} &= \frac{1}{4} & a_{12} &= \frac{1}{4} - \frac{1}{2\sqrt{3}} \\
a_{12} &= \frac{1}{4} + \frac{1}{2\sqrt{3}} & a_{22} &= \frac{1}{4}\\
c_1 &= \frac{1}{2} - \frac{1}{2\sqrt{3}} & c_2 &= \frac{1}{2} + \frac{1}{2\sqrt{3}}
\end{align}

To derive the Newton iteration, define
\begin{align}
\mathbf{z}_1 &= \mathbf{y}_n + h(a_{11}\mathbf{k}_1 + a_{12}\mathbf{k}_2) \\
\mathbf{z}_2 &= \mathbf{y}_n + h(a_{21}\mathbf{k}_1 + a_{22}\mathbf{k}_2)
\end{align}
We can then write the system of equations as
$$ \mathbf{F}(\mathbf{K}) = \left(
\begin{matrix}\mathbf{k}_1 - \mathbf{f}(t_n+c_1h,\mathbf{z}_1)\\ \mathbf{k}_2 - \mathbf{f}(t_n+c_2 h, \mathbf{z}_2)\end{matrix}\right)=0, \quad \mathrm{with}\ 
\mathbf{K} = \left(\begin{matrix}\mathbf{k}_1\\ \mathbf{k}_2\end{matrix}\right) $$
We can begin the Newton iteration with the initial guess
$$\mathbf{K}^0 = \left(\begin{matrix}\mathbf{f}(t_n,\mathbf{y}_n)\\\mathbf{f}(t_n,\mathbf{y}_n)\end{matrix}\right)$$
and proceed as
$$ \mathbf{K}^{i+1} = \mathbf{K}^i - \mathbf{d}^i$$
with $\mathbf{d}$ the solution to the linear system
$$ \mathbb{J}(\mathbf{K}^i)\ \mathbf{d}^i = \mathbf{F}(\mathbf{K}^i) $$
where $\mathbb{J}$ is the Jacobian
$$ \mathbb{J}(\mathbf{K}) = \left(
\begin{matrix}
\mathbf{I}-h a_{11}\mathbf{J}(\mathbf{z}_1) & -h a_{21}\mathbf{J}(\mathbf{z}_1) \\
-h a_{12}\mathbf{J}(\mathbf{z}_2) & \mathbf{I}-h a_{22}\mathbf{J}(\mathbf{z}_2)
\end{matrix}
\right) $$
in which $\mathbf{J}$ is the Jacobian matrix of the ODE system used previously.

```python
def implicitRK2(t, y, h, RHS, Jac, *P):
    MAXIT = 20   # raise an error after more than 20 iterations
    eps = 1e-9   # Stop iterating when the norm of the error is less than this
    
    n = y.shape[0]

    # The constants of the method
    a11 = 0.25
    a12 = 0.25 - 1/(2*np.sqrt(3))
    a21 = 0.25 + 1/(2*np.sqrt(3))
    a22 = 0.25
    c1 = 0.5 - 1/(2*np.sqrt(3))
    c2 = 0.5 + 1/(2*np.sqrt(3))
    
    #initial guess
    k1 = RHS(t,y,*P)
    k2 = k1.copy()
    
    # perform the Newton iteration
    err = 1.0
    for i in range(MAXIT):
        z1 = y + h*(a11*k1 + a12*k2)
        z2 = y + h*(a21*k1 + a22*k2)
        
        J1 = Jac(t+c1*h, z1)
        J2 = Jac(t+c2*h, z2)

        JJ11 = np.eye(n)-h*a11*J1
        JJ12 = -h*a12*J1
        JJ21 = -h*a21*J2
        JJ22 = np.eye(n)- h*a22*J2
        JJ = np.hstack(( np.vstack((JJ11,JJ21)), np.vstack((JJ12,JJ22)) ))

        F1 = k1 - RHS(t+c1*h,z1)
        F2 = k2 - RHS(t+c2*h,z2)
        F = np.vstack((F1,F2)).flatten()

        JJinv = linalg.inv(JJ)

        d = linalg.inv(JJ)@F

        # update the k's
        dd = d.reshape(2,n)
        k1 -= dd[0]
        k2 -= dd[1]
        err = np.sqrt(np.sum(d**2))
        
        if err < eps:
            return y + 0.5*h*(k1+k2)
        
    raise ValueError("too many iterations in implicitRK2")
```

```python
s = reactSys()
h = 1e-2
tRK2, yRK2 = odeSolve(s.derivs, s.jacobian, (0,100), s.y0, h, implicitRK2)
```

```python
ax = plotIt(tRK2, yRK2, sol.t, sol.y.T)
ax[0].set_title('Implicit RK2');
plt.tight_layout();
```

It is difficult to compare the results between our "exact" solution and those from implicit Euler or 
implicit RK2 since scipy's Radau stepper, unlike DOP853, has no facility to give us a solution at a constant stepsize (known as "dense output") to match our constant-stepsize odeSolve.

Instead, we can try to get an "exact" solution by using implicit RK2 with a much smaller timestep (this will take a while...)

```python
hh = 1e-4
tgood, ygood = odeSolve(s.derivs, s.jacobian, (0,100), s.y0, hh, implicitRK2)
```

First, let's make sure the solution looks reasonable

```python
ax = plotIt(tgood, ygood, sol.t, sol.y.T)
```

Now we can plot the errors of the $h=10^{-2}$ solutions

```python
def plotErr(t, y, ye):
    eps = 1e-10
    
    fig, ax = plt.subplots(2,1)
    
    ax[0].semilogy(t, np.abs(y[:,0]-ye[:,0])/np.abs(ye[:,0]+eps), label=r'$\epsilon(S_1)$')
    ax[0].semilogy(t, np.abs(y[:,1]-ye[:,1])/np.abs(ye[:,1]+eps), label=r'$\epsilon(S_2)$')
    ax[0].semilogy(t, np.abs(y[:,2]-ye[:,2])/np.abs(ye[:,2]+eps), label=r'$\epsilon(S_3)$')
    ax[0].legend(loc=1)
    ax[0].set_xlabel('t')
    ax[0].set_ylabel(r'relative error in $S_i$')

    ax[1].loglog(t, np.abs(y[:,0]-ye[:,0])/np.abs(ye[:,0]+eps),'.')
    ax[1].loglog(t, np.abs(y[:,1]-ye[:,1])/np.abs(ye[:,1]+eps),'.')
    ax[1].loglog(t, np.abs(y[:,2]-ye[:,2])/np.abs(ye[:,2]+eps),'.')
    ax[1].set_xlabel('t')
    ax[1].set_ylabel(r'relative error in $S_i$')
    
    return ax
```

```python
ax = plotErr(tE[:-1], yE[:-1], ygood[:-1:100])
ax[0].set_title('error in semi-implicit Euler')
plt.tight_layout();
```

Let's explore in more detail the ivp_solve DOP853 solution from scipy. We will use the Radau method, and then ask DOP853 to give us the solution at the points used by Radau:

```python
solR1 = solve_ivp(s.derivs, (0,100), s.y0, rtol=1e-6, method='Radau')
solDOPq= solve_ivp(s.derivs, (0,100), s.y0, t_eval=solR1.t, method='DOP853')
```

```python
ax = plotIt(solR1.t, solR1.y.T, solDOPq.t, solDOPq.y.T)
ax[0].set_title('DOP853 Dense output vs Radau')
plt.tight_layout();
```

While the solutions for $S_1$ and $S_3$ look reasonable, $S_2$ is clearly garbage. Since the DOP853 solution at the points used by the integrator looked reasonable, the flaw must arise in the dense output. DOP853 generates a polynomial representation of the solution on each interval. It then evaluates this polynomial to produce "dense output" at the requested points. Clearly the polynomials for $S_2$ are suffering from some sort of numerical problem. Plotting the errors, 

```python
ax = plotErr(solR1.t, solDOPq.y.T, solR1.y.T)
```

The Radau method we have been using is a fifth-order, fully-implicit, Runge-Kutta method with an embedded third-order method.  It uses a similar adaptive stepsize mechanism to the one we explored last lecture, and employs several "tricks" which speed up the calculation significantly. Scipy's implementation comes from E. Hairer and G. Wanner, Solving Ordinary Differential Equations II: Stiff and Differential-Algebraic Problems, starting at page 118. The book is on D2L, and is worth a glance to see what goes into such a professionally-designed algorithm.

While being expensive per timestep, on stiff systems it performs well:

```python
print(solR1.message)
print(f"number of steps: {solR1.t.shape[0]}")
print(f"number of function evaluations: {solR1.nfev}")
print(f"number of Jacobian evaluations: {solR1.njev}")
```
